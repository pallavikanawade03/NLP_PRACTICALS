```python
import spacy
from spacy import displacy

# Load English language model
nlp = spacy.load("en_core_web_sm")

# Sample sentences (can be replaced with text from a dataset)
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Alice is reading a book in the library.",
    "Machine learning is transforming the world rapidly.",
    "John gave Mary a beautiful gift.",
    "The company plans to launch a new product next year."
]

# Process and analyze each sentence
for sentence in sentences:
    print(f"\nSentence: {sentence}")
    doc = nlp(sentence)

    # Print dependency relations
    for token in doc:
        print(f"{token.text:<10} {token.dep_:<12} {token.head.text:<10} {token.pos_:<6}")

    # Visualize the syntactic structure in the browser (use only one at a time if using Jupyter)
    displacy.serve(doc, style="dep")  # comment this line if not using a browser
```
