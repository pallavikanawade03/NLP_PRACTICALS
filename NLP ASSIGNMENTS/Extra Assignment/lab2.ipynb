```python
# Import necessary libraries
import nltk
import spacy
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

# Download required NLTK packages
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

with open("/content/climate change.txt", "r", encoding="utf-8") as file:
    text = file.read()
print(text)
```

```python

tokens_nltk = word_tokenize(text)
pos_tags_nltk = pos_tag(tokens_nltk)
print("\nNLTK POS Tags:\n", pos_tags_nltk)

# POS Frequency using NLTK
pos_counts_nltk = Counter(tag for word, tag in pos_tags_nltk)
print("\nNLTK POS Frequency:\n", pos_counts_nltk)

# ---------- spaCy POS Tagging ----------

doc = nlp(text)
pos_tags_spacy = [(token.text, token.pos_) for token in doc]
print("\nspaCy POS Tags:\n", pos_tags_spacy)

# POS Frequency using spaCy
pos_counts_spacy = Counter(token.pos_ for token in doc)
print("\nspaCy POS Frequency:\n", pos_counts_spacy)
```

```python

# ---------- Visualization (Optional) ----------

# NLTK POS Frequency Plot
plt.figure(figsize=(10, 4))
plt.bar(pos_counts_nltk.keys(), pos_counts_nltk.values(), color='skyblue')
plt.title("POS Tag Frequency (NLTK)")
plt.xlabel("POS Tag")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# spaCy POS Frequency Plot
plt.figure(figsize=(10, 4))
plt.bar(pos_counts_spacy.keys(), pos_counts_spacy.values(), color='lightgreen')
plt.title("POS Tag Frequency (spaCy)")
plt.xlabel("POS Category")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
